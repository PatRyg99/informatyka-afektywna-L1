```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

```python
experiment_path = "output/10-22-2022.11:50:10"
train_predictions = os.path.join(experiment_path, "train_predictions.csv")
val_predictions = os.path.join(experiment_path, "val_predictions.csv")
```

## Emotions map

0. neutral
1. anger
2. contempt
3. disgust
4. fear
5. happy
6. sadness
7. surprise

## Training data

```python
train_df = pd.read_csv(train_predictions)

print(classification_report(train_df["true"], train_df["predicted"]))
disp = ConfusionMatrixDisplay.from_predictions(train_df["true"], train_df["predicted"])
```

## Validation data

```python
val_df = pd.read_csv(val_predictions)

print(classification_report(val_df["true"], val_df["predicted"]))
disp = ConfusionMatrixDisplay.from_predictions(val_df["true"], val_df["predicted"])
```
